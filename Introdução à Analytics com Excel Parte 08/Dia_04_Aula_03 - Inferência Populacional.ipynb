## ASCVD model

The ASCVD model has subdivisions in sex, skin color, risk in 5 or 10 years and taking or not antihypertensive medications. Since there are 2 biological sex, only 2 skin color taking in consideration (black or white), 2 year period risk and taking or not medication there are $2^4$ equations for this model. The figure below shows a table with the 16 equations for the ASCVD model. The figur was taken from https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4189930/bin/NIHMS623040-supplement-Supplemental_Material.docx.

## Módulo que cria os clusters com base em inferência estatística

```python
def create_clusters_df(n,
                       mu_age, sd_age,
                       mu_bmi, sd_bmi,
                       mu_height, sd_height,
                       mu_total_cholesterol, sd_total_cholesterol,
                       mu_hdl, sd_hdl,
                       mu_ldl, sd_ldl,
                       mu_non_hdl, sd_non_hdl,
                       mu_trygliceride, sd_trygliceride,
                       mu_fast_blood_sugar, sd_fast_blood_sugar,
                       mu_systolic_bp, sd_systolic_bp,
                       mu_diastolic_bp, sd_diastolic_bp,
                       kids0, kids1, kids2, kids3,
                       prob_kids0, prob_kids1, prob_kids2, prob_kids3,
                       p_married,p_not_married,
                       p_smokes, p_not_smoke,
                       p_drinks, p_not_drink,
                       p_family_mental_ill, p_not_family_mental_ill,
                       p_family_diabetes, p_not_family_diabetes,
                       p_family_cardiovascular, p_not_family_cardiovascular,
                       p_anxiety_episodes, p_not_anxiety_episodes,
                       prob_female, prob_male,
                       cluster_name
                      ):

    # creates gaussian distribution with mean = mu_ and standard deviation = sd_
    array_age = np.random.normal(mu_age, sd_age, n)
    array_bmi = np.random.normal(mu_bmi, sd_bmi, n)
    array_height = np.random.normal(mu_height, sd_height, n)
    array_total_cholesterol = np.random.normal(mu_total_cholesterol, sd_total_cholesterol, n)
    array_hdl = np.random.normal(mu_hdl, sd_hdl, n)
    array_ldl = np.random.normal(mu_ldl, sd_ldl, n)
    array_non_hdl = np.random.normal(mu_non_hdl, sd_non_hdl, n)
    array_trygliceride = np.random.normal(mu_trygliceride, sd_trygliceride, n)
    array_fast_blood_sugar = np.random.normal(mu_fast_blood_sugar, sd_fast_blood_sugar, n)
    array_weight = array_bmi * (array_height/100)**2 
    array_ratio_total_cholesterol_hdl = array_total_cholesterol / array_hdl 
    array_systolic_bp = np.random.normal(mu_systolic_bp, sd_systolic_bp, n)
    array_diastolic_bp = np.random.normal(mu_diastolic_bp, sd_diastolic_bp, n)


    ####################################### Random selection for given possibilities #####################################

    
    sex = [1, 2]
    prob_sex = [prob_female, prob_male]

    kids = [kids0, kids1, kids2, kids3]
    prob_kids = [prob_kids0, prob_kids1, prob_kids2, prob_kids3]

    married = [1,0]
    prob_married = [p_married,p_not_married]

    smokes = [1,0]
    prob_smokes = [p_smokes, p_not_smoke]

    drinks = [1,0]
    prob_drinks = [p_drinks, p_not_drink]

    family_mental_ilness = [1,0]
    prob_family_mental_ilness = [p_family_mental_ill, p_not_family_mental_ill]

    family_diabetes = [1,0]
    prob_family_diabetes = [p_family_diabetes, p_not_family_diabetes]

    family_cardiovascular_disease = [1,0]
    prob_family_cardiovascular_disease = [p_family_cardiovascular, p_not_family_cardiovascular]

    anxiety_episodes = [1,0]
    prob_anxiety_episodes = [p_anxiety_episodes, p_not_anxiety_episodes]

    array_sex = random.choices(population = sex, 
                               weights = prob_sex, 
                               k = n
                               )
    
    # creates array for marital status 1 = married, 0 = not married
    array_married = random.choices(population = married, 
                                   weights = prob_married, 
                                   k = n
                                  )

    # creates array for number of kids distribution
    array_kids = random.choices(population = kids, 
                                weights = prob_kids, 
                                k = n
                               )

    # creates array for number of smoker distribution
    array_smokes = random.choices(population = smokes, 
                                  weights = prob_smokes, 
                                  k = n
                                 )

    # creates array for number of drinkers distribution
    array_drinks = random.choices(population = drinks, 
                                  weights = prob_drinks, 
                                  k = n
                                 )

    # creates array for family history of mental illness distribution
    array_family_mental_ilness = random.choices(population = family_mental_ilness, 
                                                weights = prob_family_mental_ilness, 
                                                k = n
                                               )

    # creates array for family history of diabetes distribution
    array_family_diabetes = random.choices(population = family_diabetes, 
                                           weights = prob_family_diabetes, 
                                           k = n
                                          )

    # creates array for family history of heart disease distribution
    array_family_cardiovascular_disease = random.choices(population = family_cardiovascular_disease, 
                                                         weights = prob_family_cardiovascular_disease, 
                                                         k = n
                                                        )

    # creates array for anxiety episodes distribution
    array_anxiety_episodes = random.choices(population = anxiety_episodes, 
                                            weights = prob_anxiety_episodes, 
                                            k = n
                                           )

    #################################################### Creates the dataframe ############################################
    # creates arrays for each variable and sets into a list 
    data = list(zip(array_age, 
                    array_sex,
                    array_bmi, 
                    array_height, 
                    array_weight, 
                    array_kids, 
                    array_married, 
                    array_hdl, 
                    array_ldl, 
                    array_non_hdl,
                    array_total_cholesterol, 
                    array_trygliceride, 
                    array_ratio_total_cholesterol_hdl,
                    array_fast_blood_sugar, 
                    array_systolic_bp,
                    array_diastolic_bp,
                    array_smokes, 
                    array_drinks, 
                    array_family_mental_ilness, 
                    array_family_diabetes,
                    array_family_cardiovascular_disease,
                    array_anxiety_episodes
                   ))

    # create the heading for the dataframe df01
    columns = ['age_years',
               'sex',
               'bmi_kg_m2',
               'height_cm',
               'weight_kg',
               'number_of_children',
               'flag_married',
               'hdl_mg_dl',
               'ldl_mg_dl',
               'non_hdl_mg_dl',
               'total_cholesterol_mg_dl',
               'trygliceride_mg_dl', 
               'ratio_total_cholesterol_hdl',
               'fast_blood_sugar_mg_dl',
               'systolic_bp_mm_hg',
               'diastolic_bp_mm_hg',
               'flag_smokes',
               'flag_drinks', 
               'flag_family_mental_ilness',
               'flag_family_diabetes',
               'flag_family_cardiovascular_disease',
               'flag_anxiety_episodes'
              ]

    flag_features = ['flag_smokes',
                     'flag_drinks', 
                     'flag_family_mental_ilness',
                     'flag_family_diabetes',
                     'flag_family_cardiovascular_disease',
                     'flag_anxiety_episodes'    
                    ]

    df = pd.DataFrame(data, columns = columns) 

    ########################################## Calculations and other features ############################################

    df['flag_sex'] = df.apply(sex_variable, axis=1)

    df['cluster'] = cluster_name
    df['carhill_score'] = df.apply(score_carhill,axis=1)

    df['sum_risk_flag_1'] = df[flag_features].sum(axis=1)
    df['index_risk_flag_1'] = df[flag_features].sum(axis=1)/len(flag_features)

    df['index_bmi'] = df.apply(bmi, axis=1)
    df['index_diabetes'] = df.apply(fasting_diabetes, axis=1)
    df['index_total_cholesterol'] = df.apply(total_cholesterol, axis=1)
    df['index_hdl_cholesterol'] = df.apply(hdl_cholesterol, axis=1)
    df['index_non_hdl_cholesterol'] = df.apply(non_hdl_cholesterol, axis=1)
    df['index_ldl_cholesterol'] = df.apply(ldl_cholesterol, axis=1)
    df['index_tryglicerides'] = df.apply(tryglicerides, axis=1)
    df['index_ratio_total_cholest_hdl'] = df.apply(ratio_total_cholest_hdl, axis=1)
    df['index_systolic_blood_pressure'] = df.apply(systolic_blood_pressure, axis=1)
    df['index_diastolic_blood_pressure'] = df.apply(diastolic_blood_pressure, axis=1)
    df['index_blood_pressure'] = df.apply(blood_pressure, axis=1)
    
    index_features = ['index_bmi',
                      'index_diabetes',
                      'index_total_cholesterol',
                      'index_hdl_cholesterol', 
                      'index_non_hdl_cholesterol',
                      'index_ldl_cholesterol',
                      'index_tryglicerides',
                      'index_ratio_total_cholest_hdl',
                      'index_systolic_blood_pressure',
                      'index_diastolic_blood_pressure',
                      'index_blood_pressure'    
                    ]
    
    df['sum_risk_flag_2'] = df[index_features].sum(axis=1)
    df['index_risk_flag_2'] = df[index_features].sum(axis=1)/len(index_features)
    
    df['target_risk_flag_1'] = df.apply(risk_index_func_1, axis=1)
    df['target_risk_flag_2'] = df.apply(risk_index_func_2, axis=1)      
    df['target_risk_flag_3'] = df.apply(risk_index_func_3, axis=1)   
    
    df.reset_index(drop=True)
    df
    return df
```

# Simulando a evolução no tempo das features do modelo

## Método que simula as features via passeio aleatório

```python
def time_series(max, n, a, b, c, date_start):
    stop = int((max - b)/a)
    x = np.linspace(0, stop, n)
    y = a * x + b + c * np.random.uniform(low = -15, high = 15, size = n) # np.random.normal(0,1,n)
    df_test = pd.DataFrame(y,columns=['name']).reset_index(drop=True)
    df_test['date'] = pd.date_range(start = date_start, periods = n)
    #fig, ax = plt.subplots()
    #ax.plot(x, y, label = "Data")
    return df_test
```

## Método que cria dataframe para o ASCVD

```python
def create_clusters_df(n,
                       mu_age, sd_age,
                       prob_white, prob_black,
                       mu_bmi, sd_bmi,
                       mu_height, sd_height,
                       mu_total_cholesterol, sd_total_cholesterol,
                       mu_hdl, sd_hdl,
                       mu_ldl, sd_ldl,
                       mu_non_hdl, sd_non_hdl,
                       mu_trygliceride, sd_trygliceride,
                       mu_fast_blood_sugar, sd_fast_blood_sugar,
                       mu_systolic_bp, sd_systolic_bp,
                       mu_diastolic_bp, sd_diastolic_bp,
                       kids0, kids1, kids2, kids3,
                       prob_kids0, prob_kids1, prob_kids2, prob_kids3,
                       p_married,p_not_married,
                       p_smokes, p_not_smoke,
                       p_drinks, p_not_drink,
                       p_family_mental_ill, p_not_family_mental_ill,
                       p_family_diabetes, p_not_family_diabetes,
                       p_family_cardiovascular, p_not_family_cardiovascular,
                       p_anxiety_episodes, p_not_anxiety_episodes,
                       prob_female, prob_male,
                       p_antihyp_meds,
                       cluster_name
                      ):

    # creates gaussian distribution with mean = mu_ and standard deviation = sd_
    array_age = np.random.normal(mu_age, sd_age, n)
    array_bmi = np.random.normal(mu_bmi, sd_bmi, n)
    array_height = np.random.normal(mu_height, sd_height, n)
    array_total_cholesterol = np.random.normal(mu_total_cholesterol, sd_total_cholesterol, n)
    array_hdl = np.random.normal(mu_hdl, sd_hdl, n)
    array_ldl = np.random.normal(mu_ldl, sd_ldl, n)
    array_non_hdl = np.random.normal(mu_non_hdl, sd_non_hdl, n)
    array_trygliceride = np.random.normal(mu_trygliceride, sd_trygliceride, n)
    array_fast_blood_sugar = np.random.normal(mu_fast_blood_sugar, sd_fast_blood_sugar, n)
    array_weight = array_bmi * (array_height/100)**2 
    array_ratio_total_cholesterol_hdl = array_total_cholesterol / array_hdl 
    array_systolic_bp = np.random.normal(mu_systolic_bp, sd_systolic_bp, n)
    array_diastolic_bp = np.random.normal(mu_diastolic_bp, sd_diastolic_bp, n)


    ####################################### Random selection for given possibilities #####################################

    
    sex = ['F', 'M']
    prob_sex = [prob_female, prob_male]
    
    skin_color = ['W', 'B']
    prob_skin_color = [prob_white, prob_black]

    kids = [kids0, kids1, kids2, kids3]
    prob_kids = [prob_kids0, prob_kids1, prob_kids2, prob_kids3]

    married = [1,0]
    prob_married = [p_married,p_not_married]

    smokes = [1,0]
    prob_smokes = [p_smokes, p_not_smoke]

    drinks = [1,0]
    prob_drinks = [p_drinks, p_not_drink]

    family_mental_ilness = [1,0]
    prob_family_mental_ilness = [p_family_mental_ill, p_not_family_mental_ill]

    family_diabetes = [1,0]
    prob_family_diabetes = [p_family_diabetes, p_not_family_diabetes]

    family_cardiovascular_disease = [1,0]
    prob_family_cardiovascular_disease = [p_family_cardiovascular, p_not_family_cardiovascular]

    anxiety_episodes = [1,0]
    prob_anxiety_episodes = [p_anxiety_episodes, p_not_anxiety_episodes]
    
    antihyp_meds = [1,0]
    prob_antihyp_meds = [p_antihyp_meds, 1 - p_antihyp_meds]
    
    
    array_skin_color = random.choices(population = skin_color, 
                               weights = prob_skin_color, 
                               k = n
                               )    

    array_sex = random.choices(population = sex, 
                               weights = prob_sex, 
                               k = n
                               )
    
    # creates array for marital status 1 = married, 0 = not married
    array_married = random.choices(population = married, 
                                   weights = prob_married, 
                                   k = n
                                  )

    # creates array for number of kids distribution
    array_kids = random.choices(population = kids, 
                                weights = prob_kids, 
                                k = n
                               )

    # creates array for number of smoker distribution
    array_smokes = random.choices(population = smokes, 
                                  weights = prob_smokes, 
                                  k = n
                                 )

    # creates array for number of drinkers distribution
    array_drinks = random.choices(population = drinks, 
                                  weights = prob_drinks, 
                                  k = n
                                 )

    # creates array for family history of mental illness distribution
    array_family_mental_ilness = random.choices(population = family_mental_ilness, 
                                                weights = prob_family_mental_ilness, 
                                                k = n
                                               )

    # creates array for family history of diabetes distribution
    array_family_diabetes = random.choices(population = family_diabetes, 
                                           weights = prob_family_diabetes, 
                                           k = n
                                          )

    # creates array for family history of heart disease distribution
    array_family_cardiovascular_disease = random.choices(population = family_cardiovascular_disease, 
                                                         weights = prob_family_cardiovascular_disease, 
                                                         k = n
                                                        )

    # creates array for anxiety episodes distribution
    array_anxiety_episodes = random.choices(population = anxiety_episodes, 
                                            weights = prob_anxiety_episodes, 
                                            k = n
                                           )
    
    # creates array for taking antihypertensive medications distribution
    array_antihyp_meds = random.choices(population = antihyp_meds, 
                                            weights = prob_antihyp_meds, 
                                            k = n
                                           )

    #################################################### Creates the dataframe ############################################
    # creates arrays for each variable and sets into a list 
    data = list(zip(array_age, 
                    array_skin_color,
                    array_sex,
                    array_bmi, 
                    array_height, 
                    array_weight, 
                    array_kids, 
                    array_married, 
                    array_hdl, 
                    array_ldl, 
                    array_non_hdl,
                    array_total_cholesterol, 
                    array_trygliceride, 
                    array_ratio_total_cholesterol_hdl,
                    array_fast_blood_sugar, 
                    array_systolic_bp,
                    array_diastolic_bp,
                    array_smokes, 
                    array_drinks, 
                    array_family_mental_ilness, 
                    array_family_diabetes,
                    array_family_cardiovascular_disease,
                    array_anxiety_episodes,
                    array_antihyp_meds
                   ))

    # create the heading for the dataframe
    columns = ['age_years',
               'skin_color',
               'sex',
               'bmi_kg_m2',
               'height_cm',
               'weight_kg',
               'number_of_children',
               'flag_married',
               'hdl_mg_dl',
               'ldl_mg_dl',
               'non_hdl_mg_dl',
               'total_cholesterol_mg_dl',
               'trygliceride_mg_dl', 
               'ratio_total_cholesterol_hdl',
               'fast_blood_sugar_mg_dl',
               'systolic_bp_mm_hg',
               'diastolic_bp_mm_hg',
               'flag_smokes',
               'flag_drinks', 
               'flag_family_mental_ilness',
               'flag_family_diabetes',
               'flag_family_cardiovascular_disease',
               'flag_anxiety_episodes',
               'flag_antihyp_meds'
              ]

    flag_features = ['flag_smokes',
                     'flag_drinks', 
                     'flag_family_mental_ilness',
                     'flag_family_diabetes',
                     'flag_family_cardiovascular_disease',
                     'flag_anxiety_episodes'    
                    ]

    df = pd.DataFrame(data, columns = columns) 

    ########################################## Calculations and other features ############################################

    df['flag_sex'] = df.apply(sex_variable, axis=1)

    df['cluster'] = cluster_name
    df['carhill_score'] = df.apply(score_carhill,axis=1)

    df['sum_risk_flag_1'] = df[flag_features].sum(axis=1)
    df['index_risk_flag_1'] = df[flag_features].sum(axis=1)/len(flag_features)

    df['index_bmi'] = df.apply(bmi, axis=1)
    df['index_diabetes'] = df.apply(fasting_diabetes, axis=1)
    df['index_total_cholesterol'] = df.apply(total_cholesterol, axis=1)
    df['index_hdl_cholesterol'] = df.apply(hdl_cholesterol, axis=1)
    df['index_non_hdl_cholesterol'] = df.apply(non_hdl_cholesterol, axis=1)
    df['index_ldl_cholesterol'] = df.apply(ldl_cholesterol, axis=1)
    df['index_tryglicerides'] = df.apply(tryglicerides, axis=1)
    df['index_ratio_total_cholest_hdl'] = df.apply(ratio_total_cholest_hdl, axis=1)
    df['index_systolic_blood_pressure'] = df.apply(systolic_blood_pressure, axis=1)
    df['index_diastolic_blood_pressure'] = df.apply(diastolic_blood_pressure, axis=1)
    df['index_blood_pressure'] = df.apply(blood_pressure, axis=1)
    
    index_features = ['index_bmi',
                      'index_diabetes',
                      'index_total_cholesterol',
                      'index_hdl_cholesterol', 
                      'index_non_hdl_cholesterol',
                      'index_ldl_cholesterol',
                      'index_tryglicerides',
                      'index_ratio_total_cholest_hdl',
                      'index_systolic_blood_pressure',
                      'index_diastolic_blood_pressure',
                      'index_blood_pressure'    
                    ]
    
    df['sum_risk_flag_2'] = df[index_features].sum(axis=1)
    df['index_risk_flag_2'] = df[index_features].sum(axis=1)/len(index_features)
    
    df['target_risk_flag_1'] = df.apply(risk_index_func_1, axis=1)
    df['target_risk_flag_2'] = df.apply(risk_index_func_2, axis=1)      
    df['target_risk_flag_3'] = df.apply(risk_index_func_3, axis=1)   
    
    df.reset_index(drop=True)
    df
    return df
```

## Black women not taking antihypertensive medications

```python
# Calculates the individual score 
df01['logit_ascvd'] = 17.114 * df01['ln_age'] 
+ 0.94 * df01['ln_total_cholesterol_mg_dl'] 
- 18.92 * df01['ln_hdl_mg_dl'] 
+ 4.475 * df01['ln_age'] * df01['ln_hdl_mg_dl'] 
+ 27.82 * df01['ln_systolic_bp_mm_hg'] 
- 6.087 * df01['ln_age'] * df01['ln_systolic_bp_mm_hg'] \
* 0.691 * df01['flag_smokes'] * 0.874 * df01['flag_diabetes']

# S0 value for 5 years score calculation
s05y_df01 = 0.98194

# S0 value for 10 years score calculation
s010y_df01 = 0.98194

# Mean score value for individual score calculation
ms_df01 = 86.61

# Individual score calculation for 5 years risk assessment
df01['score_5years'] = 1 - s05y_df01 * np.exp(df01['logit_ascvd'] - ms_df01)

# Individual score calculation for 10 years risk assessment
df01['score_10years'] = 1 - s010y_df01 * np.exp(df01['logit_ascvd'] - ms_df01)

df01.head()
```
